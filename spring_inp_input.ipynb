```python
def process_inp_file(file_path):  
    # 读取原始文件  
    with open(file_path, 'r') as file:  
        content = file.readlines()  
    
    # 找到实例定义的位置  
    instance_start_index = -1  
    for i, line in enumerate(content):  
        if '*Instance, name=PART-1-1, part=PART-1' in line:  
            instance_start_index = i  
            break  
    
    if instance_start_index == -1:  
        print("未找到指定实例")  
        return None  
    
    # 角点节点  
    corner_nodes = [1, 61, 733, 793]  
    corner_elements = []  
    for i, node in enumerate(corner_nodes, start=1001):  
        corner_elements.append(f"*Element, type=SPRING1, elset=CornerSprings\n{i}, {node}\n")  
    
    # 边界节点  
    side_nodes_groups = [  
        (2, 61, 1),    # 底部左侧  
        (734, 793, 1), # 底部右侧  
        (62, 733, 61),  # 左侧  
        (122, 793, 61)  # 右侧  
    ]  
    side_elements = []  
    element_counter = 1005  
    for start, end, direction in side_nodes_groups:  
        for node in range(start, end, direction):  
            side_elements.append(f"*Element, type=SPRING1, elset=SideSprings\n{element_counter}, {node}\n")  
            element_counter += 1  
    
    # 中间节点  
    middle_nodes_ranges = [  
        (63, 122,1),  
        (124, 183,1),  
        (185, 244,1),  
        (246, 305,1),  
        (307, 366,1),  
        (368, 427,1),  
        (429, 488,1),  
        (490, 549,1),  
        (551, 610,1),  
        (612, 671,1),  
        (673, 732,1)  
    ]  
    middle_elements = []  
    element_counter = 1009  
    for  start, end, direction  in middle_nodes_ranges:  
        for node in range(start, end, direction):  
            middle_elements.append(f"*Element, type=SPRING1, elset=MiddleSprings\n{element_counter}, {node}\n")  
            element_counter += 1  
    
    # 组合所有单元  
    spring_elements = corner_elements + side_elements + middle_elements  
    
    # 在原文件中插入弹簧单元定义  
    content[instance_start_index+1:instance_start_index+1] = spring_elements  
    
    # 在弹簧单元之后添加刚度定义  
    spring_stiffness_lines = [  
        "*Spring, elset=CornerSprings\n",  
        "61312.5, 3\n",  
        "*Spring, elset=SideSprings\n",  
        "122625, 3\n",   
        "*Spring, elset=MiddleSprings\n",  
        "245250, 3\n"  
    ]  
    content[instance_start_index+1+len(spring_elements):instance_start_index+1+len(spring_elements)] = spring_stiffness_lines  
    
    # 写入新文件  
    output_file_path = file_path.replace('.inp', '_modified.inp')  
    with open(output_file_path, 'w') as file:  
        file.writelines(content)  
    
    print(f"Modified file saved as {output_file_path}")  
    return output_file_path  

# 使用示例  
file_path = r'E:\phd\Code\DM-FEM2D\FEM_Reduce\RODM_Ad_data\valiation_force\RODM_force_standard_ad_serep_3_spring.inp'  
process_inp_file(file_path)
```
