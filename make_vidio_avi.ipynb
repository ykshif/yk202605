```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os

# 加载复数位移数据
heave_1 = np.load('heave_3.npy')  # 假设 heave_1 是一个 (13, 61) 的复数数组

# 提取数据的模值和相位
magnitude_data = np.abs(heave_1)
phase_data = np.angle(heave_1)  # 提取相位信息

# 设置视频帧大小为480p
frame_height = 480
frame_width = 854

duration = 20  # 生成20秒的视频
fps = 60  # 每秒60帧
num_frames = duration * fps  # 总帧数
omega = 1.1  # 角频率0.8 rad/s

# 设置视频保存路径和名称，保存为 avi 格式，使用 MJPG 编码
output_video_path = 'heave_displacement_video.avi'
fourcc = cv2.VideoWriter_fourcc(*'MJPG')
out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

# 创建保存每一帧图像的文件夹
if not os.path.exists('frames'):
    os.makedirs('frames')

# 生成每一帧图像并保存到视频中
for i in range(num_frames):
    # 根据角频率和帧数生成当前时刻的复合响应
    time = i / fps
    current_frame_data = magnitude_data * np.sin(omega * time + phase_data)

    # 绘制当前帧
    plt.figure(figsize=(frame_width / 100, frame_height / 100))  # 设置图像尺寸，以适应480p分辨率
    plt.imshow(current_frame_data,cmap= 'RdYlBu_r',interpolation='spline16',extent=(0,300,0,60))  # 用颜色表示幅值与相位的复合结果
    #plt.colorbar(label='Combined Heave Response')  # 颜色条表示合成后的响应值
    plt.title(f'Time {time:.2f}s')
    plt.xlabel("X")  # 设置横坐标标签
    plt.ylabel("Y")  # 设置纵坐标标签
    plt.savefig(f'frames/frame_{i}.png', dpi=300)  # 设置高 dpi 以提高清晰度
    plt.close()

    # 将图像转换为视频帧
    frame = cv2.imread(f'frames/frame_{i}.png')
    frame = cv2.resize(frame, (frame_width, frame_height))  # 确保帧大小为480p
    out.write(frame)

# 释放视频资源
out.release()

# 清除临时生成的图像帧
for i in range(num_frames):
    os.remove(f'frames/frame_{i}.png')

os.rmdir('frames')

print("视频生成完成，已保存为 heave_displacement_video1.avi")
```
